```python
# ==============================================================================
# F1 2025 LAS VEGAS GP PREDICTOR - HIGHLY REALISTIC COLD WEATHER SIMULATION
# ==============================================================================
# This simulation incorporates specific user requests:
# 1. Roster Update: Tsunoda (TSU) to Red Bull Racing, Lawson (LAW) to RB.
# 2. Performance Hypothesis: Vegas cold temperatures favor Red Bull and Mercedes
#    (due to better tire warm-up), while penalizing McLaren.
# 3. Rookie Performance: Penalized based on their car's pace.
# ==============================================================================

import fastf1
import os
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Suppress noisy warnings and set display options
warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
pd.options.mode.chained_assignment = None
pd.set_option('display.max_rows', 10)

# ==============================================================================
# SECTION 1: SETUP AND DATA PREPARATION
# ==============================================================================

class DataManager:
    """Handles FastF1 API connections and Caching."""
    def __init__(self, cache_dir='f1_cache_vegas'):
        self.cache_dir = cache_dir
        self._setup_cache()

    def _setup_cache(self):
        """Sets up the FastF1 caching directory."""
        if not os.path.exists(self.cache_dir):
            os.makedirs(self.cache_dir)
        fastf1.Cache.enable_cache(self.cache_dir)
        print(f"[Setup] FastF1 Cache enabled at: {os.path.abspath(self.cache_dir)}")

    def get_session_data(self, year, gp_name, session_type='R', load_weather=False):
        """Loads F1 session data from FastF1."""
        print(f"[Data] Loading {year} {gp_name} - Session: {session_type}...")
        try:
            # We specifically load the 2023 Las Vegas data as the baseline
            session = fastf1.get_session(year, gp_name, session_type)
            session.load(weather=load_weather)
            return session
        except Exception as e:
            print(f"[Warning] Could not load session {year} {gp_name}: {e}. Skipping...")
            return None

# ==============================================================================
# SECTION 2: MACHINE LEARNING PREDICTION FRAMEWORK
# ==============================================================================

class FeatureEngineer:
    """Processes raw lap data into machine learning features."""
    def __init__(self):
        self.le_driver = LabelEncoder()
        self.le_compound = LabelEncoder()
        # Initialize a persistent driver encoder for 2025 prediction mapping
        self.driver_map = {} 

    def preprocess_data(self, race_session, quali_session=None, prev_year_session=None):
        """Extracts and engineers features from the loaded session data."""
        if not hasattr(race_session, 'laps'): return None
        
        print(f"[Features] Processing {race_session.event.year} data for training...")
        laps = race_session.laps
        
        # Filter for valid, accurate racing laps
        laps = laps[laps['LapTime'].notna()].reset_index(drop=True)
        if 'IsAccurate' in laps.columns:
            laps = laps[laps['IsAccurate'] == True]
        
        # --- DEPENDENT VARIABLE (Target) ---
        laps['LapTime_Seconds'] = laps['LapTime'].dt.total_seconds()

        # --- FEATURES ---
        laps['TireAge'] = laps['TyreLife']
        laps['Compound_Enc'] = self.le_compound.fit_transform(laps['Compound'])
        
        # Cold Weather Sensitive Team Strength (Crucial Feature)
        laps['TeamStrength'] = laps['Team'].apply(self._get_cold_weather_team_strength)

        # Clean Air Pace (Top 3 avg laps)
        clean_air_pace = laps.groupby('Driver')['LapTime'].apply(
            lambda x: x.nsmallest(3).mean()
        ).reset_index().rename(columns={'LapTime': 'CleanAirPace'})
        laps = laps.merge(clean_air_pace, on='Driver', how='left')
        laps['CleanAirPace_Sec'] = laps['CleanAirPace'].dt.total_seconds()

        # Qualifying Pace (Crucial for prediction bias)
        if quali_session:
            q_times = self._extract_quali_pace(quali_session)
            laps = laps.merge(q_times, on='Driver', how='left')
        else:
            laps['QualiTime'] = pd.NaT
        
        # Previous Year Pace (History feature)
        if prev_year_session:
            hist_pace = self._extract_historical_pace(prev_year_session)
            laps = laps.merge(hist_pace, on='Driver', how='left')
        else:
            laps['PrevYearAvg'] = pd.NaT

        # Finalizing Feature Columns
        for col in ['QualiTime', 'PrevYearAvg']:
            if col in laps.columns:
                laps[f'{col}_Sec'] = laps[col].dt.total_seconds()
            else:
                laps[f'{col}_Sec'] = 0.0
        
        # Fill NaNs with cautious estimates
        laps['PrevYearAvg_Sec'] = laps['PrevYearAvg_Sec'].fillna(laps['LapTime_Seconds'].mean())
        laps['QualiTime_Sec'] = laps['QualiTime_Sec'].fillna(laps.LapTime_Seconds.min() * 1.05)
        
        # Encode drivers and store mapping
        laps['Driver_Enc'] = self.le_driver.fit_transform(laps['Driver'])
        self.driver_map = dict(zip(self.le_driver.classes_, self.le_driver.transform(self.le_driver.classes_)))

        return laps.dropna(subset=['LapTime_Seconds'])

    def _get_cold_weather_team_strength(self, team):
        """
        Assigns team strength based on the Cold Vegas temperature hypothesis.
        Higher number = stronger car/performance.
        """
        ratings = {
            'Red Bull Racing': 0.98,  # Dominant pace, cold temps are fine.
            'Mercedes': 0.96,         # Boosted due to better cold tire warm-up historically.
            'Ferrari': 0.95,          # Very strong, but slightly behind RBR/Merc in the cold here.
            'McLaren': 0.90,          # Penalized significantly due to cold track/tire issues.
            'Aston Martin': 0.70,     
            'Haas': 0.65,             # Solid midfield.
            'RB': 0.60,
            'Williams': 0.58, 
            'Alpine': 0.55, 
            'Kick Sauber': 0.40
        }
        return ratings.get(team, 0.5)

    def _extract_quali_pace(self, session):
        """Extracts the best qualifying time per driver."""
        try:
            best = session.laps.pick_fastest_per_driver()
        except:
            best = session.laps.groupby('Driver').first().reset_index()
        return best[['Driver', 'LapTime']].rename(columns={'LapTime': 'QualiTime'})

    def _extract_historical_pace(self, session):
        """Extracts the average lap time per driver from a session."""
        valid = session.laps[session.laps['LapTime'].notna()]
        if 'IsAccurate' in valid.columns: valid = valid[valid['IsAccurate']==True]
        avg = valid.groupby('Driver')['LapTime'].mean().reset_index()
        return avg.rename(columns={'LapTime': 'PrevYearAvg'})

class LapTimePredictor:
    """Machine Learning model for lap time prediction."""
    def __init__(self):
        self.model = RandomForestRegressor(n_estimators=100, random_state=2025)
        self.features = [
            'Driver_Enc', 'Compound_Enc', 'TireAge', 'TeamStrength', 
            'CleanAirPace_Sec', 'QualiTime_Sec', 'PrevYearAvg_Sec'
        ]
        self.feature_engineer = None

    def train(self, df, fe):
        """Trains the Random Forest model."""
        self.feature_engineer = fe
        X = df[self.features]
        y = df['LapTime_Seconds']
        print(f"[Model] Training on {len(X)} historical laps.")
        self.model.fit(X, y)

    def predict_2025(self, input_df):
        """Makes predictions for the 2025 grid."""
        # Ensure new drivers/teams are encoded correctly using the existing map
        grid_df = input_df.copy()
        
        def safe_encode_driver(driver):
            return self.feature_engineer.driver_map.get(driver, 0) # Fallback to 0 if new driver

        grid_df['Driver_Enc'] = grid_df['Driver'].apply(safe_encode_driver)
        
        # Compound_Enc: Assuming all start on the same compound (e.g., Hard/C2 encoded as 2)
        grid_df['Compound_Enc'] = self.feature_engineer.le_compound.transform(['HARD'])[0] 
        
        X = grid_df[self.features]
        return self.model.predict(X)

# ==============================================================================
# SECTION 3: SIMULATION (REALISTIC 2025 GRID)
# ==============================================================================

class SimulationEngine:
    """Constructs the 2025 F1 grid with performance adjustments."""
    def create_manual_2025_grid(self, base_df):
        """
        Creates a DataFrame representing the 2025 grid with specific driver/team transfers.
        Performance features are synthesized using a mix of team pace and driver history.
        """
        print("[Simulation] Constructing Realistic 2025 Grid...")
        
        # 2025 CONFIRMED/RUMORED ROSTER (Incorporating user's specific swaps)
        roster_2025 = {
            'VER': 'Red Bull Racing', 'TSU': 'Red Bull Racing', # Tsunoda Transfer
            'HAM': 'Ferrari',         'LEC': 'Ferrari',         # Hamilton Transfer
            'RUS': 'Mercedes',        'ANT': 'Mercedes',        # Antonelli Rookie
            'NOR': 'McLaren',         'PIA': 'McLaren',
            'ALO': 'Aston Martin',    'STR': 'Aston Martin',
            'GAS': 'Alpine',          'DOO': 'Alpine',          # Doohan Rookie
            'ALB': 'Williams',        'SAI': 'Williams',        # Sainz Transfer
            'OCO': 'Haas',            'BEA': 'Haas',            # Ocon Transfer / Bearman Rookie
            'LAW': 'RB',              'HAD': 'RB',              # Lawson Transfer (from RBR to RB) / Hadjar Rookie
            'HUL': 'Kick Sauber',     'BOR': 'Kick Sauber',     # Hulk Transfer / Bortoleto Rookie
        }

        # Calculate mean performance metrics from the 2023 training data
        driver_stats_2024 = base_df.groupby('Driver').first()
        team_stats_2024 = base_df.groupby('Team').mean(numeric_only=True)
        
        grid_rows = []

        # Define performance penalties/bonuses in seconds
        ROOKIE_PENALTY = 0.45  # Consistent performance penalty for rookies
        TRANSFER_BONUS = -0.10 # Slight boost for proven drivers settling in new teams

        for driver, team in roster_2025.items():
            row = {'Driver': driver, 'Team': team, 'TireAge': 1.0}
            
            # 1. Get the Target Team's Base Performance (Mechanical Pace)
            team_perf = team_stats_2024.loc[team] if team in team_stats_2024.index else team_stats_2024.mean() 

            # 2. Get Driver's Personal Skill (History)
            driver_history = driver_stats_2024.loc[driver]['PrevYearAvg_Sec'] if driver in driver_stats_2024.index else None

            # --- APPLY LOGIC BASED ON DRIVER STATUS ---
            
            is_rookie = driver in ['ANT', 'BEA', 'DOO', 'BOR', 'HAD']
            
            if is_rookie:
                # ROOKIES: Performance = New Team's Speed + Rookie Penalty (0.45s)
                # This accurately captures that the car is the limiting factor + driver inexperience
                row['QualiTime_Sec'] = team_perf['QualiTime_Sec'] + ROOKIE_PENALTY
                row['CleanAirPace_Sec'] = team_perf['CleanAirPace_Sec'] + ROOKIE_PENALTY
                row['PrevYearAvg_Sec'] = team_perf['PrevYearAvg_Sec'] + ROOKIE_PENALTY
                
            elif driver in ['HAM', 'SAI', 'OCO', 'HUL', 'TSU', 'LAW']:
                # TRANSFER/TRANSITION DRIVERS:
                # Performance = New Team's Speed (Car Strength) + Driver Consistency (History)
                
                # Base performance is the new team's expected pace
                row['QualiTime_Sec'] = team_perf['QualiTime_Sec']
                row['CleanAirPace_Sec'] = team_perf['CleanAirPace_Sec']
                row['PrevYearAvg_Sec'] = team_perf['PrevYearAvg_Sec']
                
                # Apply driver consistency/skill boost if historical pace is available
                if driver_history:
                    # Blend: 70% Team Pace (Car) + 30% Driver History (Skill)
                    row['PrevYearAvg_Sec'] = (row['PrevYearAvg_Sec'] * 0.7) + (driver_history * 0.3)
                    
                row['PrevYearAvg_Sec'] += TRANSFER_BONUS # Give a slight bonus for being an established driver
                
            else:
                # ESTABLISHED DRIVERS (VER, LEC, RUS, NOR, etc.):
                # Performance = Their 2024 Stats with a slight development gain
                if driver in driver_stats_2024.index:
                    stats = driver_stats_2024.loc[driver]
                    row['QualiTime_Sec'] = stats['QualiTime_Sec'] - 0.15 # Development gain
                    row['CleanAirPace_Sec'] = stats['CleanAirPace_Sec'] - 0.10
                    row['PrevYearAvg_Sec'] = stats['PrevYearAvg_Sec'] - 0.05
                else:
                    # Fallback to team average if 2024 data missing
                    row['QualiTime_Sec'] = team_perf['QualiTime_Sec']
                    row['CleanAirPace_Sec'] = team_perf['CleanAirPace_Sec']
                    row['PrevYearAvg_Sec'] = team_perf['PrevYearAvg_Sec']

            # Set the Team Strength feature based on the NEW team's cold-weather rating
            row['TeamStrength'] = FeatureEngineer()._get_cold_weather_team_strength(team)
            
            grid_rows.append(row)

        return pd.DataFrame(grid_rows)

class Visualizer:
    """Plots the simulation results."""
    def plot_2025_predictions(self, result_df):
        sns.set_style("whitegrid")
        plt.figure(figsize=(14, 8))
        
        result_df = result_df.sort_values('Predicted_LapTime_2025', ascending=False) # Plot fastest at the top
        
        # Color Coding based on Team Identity
        team_colors = {
            'Red Bull Racing': '#0600ef', 'Ferrari': '#dc0000',
            'Mercedes': '#00d2be', 'McLaren': '#ff8700',
            'Aston Martin': '#006f62', 'Alpine': '#0090ff',
            'Williams': '#005aff', 'Haas': '#ffffff',
            'Kick Sauber': '#52e252', 'RB': '#6692ff'
        }
        
        colors = [team_colors.get(team, '#333333') for team in result_df['Team']]
        
        ax = sns.barplot(
            data=result_df, x='Predicted_LapTime_2025', y='Driver', 
            palette=colors, edgecolor='black'
        )
        
        # Dynamic X-Axis Limits for better visualization
        min_time = result_df['Predicted_LapTime_2025'].min()
        max_time = result_df['Predicted_LapTime_2025'].max()
        plt.xlim(min_time - 0.5, max_time + 0.5)
        
        plt.title('2025 Las Vegas GP Prediction (Cold Track, Realistic Transfers)', fontsize=16)
        plt.xlabel('Predicted Race Pace (Lap Time in Seconds)', fontsize=12)
        
        # Add Team and Predicted Time to the bars
        for i, (p, team, time) in enumerate(zip(ax.patches, result_df['Team'], result_df['Predicted_LapTime_2025'])):
            ax.text(p.get_x() + p.get_width() + 0.05, p.get_y() + p.get_height()/2, 
                    f"{team} ({time:.3f}s)", va='center', fontsize=9, color='black')
        
        plt.show()

# ==============================================================================
# MAIN ORCHESTRATOR
# ==============================================================================

def main():
    print("--- F1 2025 Las Vegas GP Predictor (Highly Realistic Cold Edition) ---")
    
    dm = DataManager()
    fe = FeatureEngineer()
    predictor = LapTimePredictor()
    sim = SimulationEngine()
    viz = Visualizer()

    # 1. Data Pull - STRICTLY LAS VEGAS 2023 for track profile training
    GP_NAME = 'Las Vegas' 
    print(f"\n[Step 1] Fetching Historical Data for {GP_NAME}...")
    
    race_data = dm.get_session_data(2023, GP_NAME, 'R')
    quali_data = dm.get_session_data(2023, GP_NAME, 'Q')
    
    if race_data is None or quali_data is None:
        print("Critical Error: Las Vegas 2023 data unavailable. Cannot run simulation.")
        return

    # 2. Train Model
    print("\n[Step 2] Training Model on Las Vegas 2023 Data...")
    train_df = fe.preprocess_data(race_data, quali_data, race_data) 
    if train_df is None or train_df.empty:
        print("Error: Training data is empty after processing.")
        return
        
    predictor.train(train_df, fe)

    # 3. Simulate 2025 Grid
    print("\n[Step 3] Constructing Realistic 2025 Grid with Cold-Weather Bias...")
    grid_2025 = sim.create_manual_2025_grid(train_df)
    
    # 4. Predict & Output
    print("[Step 4] Predicting Results...")
    grid_2025['Predicted_LapTime_2025'] = predictor.predict_2025(grid_2025)
    
    # Format and Print Results
    print("\n" + "="*70)
    print("2025 LAS VEGAS GP PREDICTED CLASSIFICATION (COLD TRACK BIAS)")
    print("="*70)
    
    final_results = grid_2025[['Driver', 'Team', 'Predicted_LapTime_2025']] \
        .sort_values('Predicted_LapTime_2025').reset_index(drop=True)
    
    final_results['Predicted_LapTime_2025'] = final_results['Predicted_LapTime_2025'].apply(lambda x: f"{x:.3f}s")
    final_results.index = final_results.index + 1
    final_results.index.name = 'Pos'
    
    print(final_results)
    print("="*70 + "\n")

    # 5. Visualize
    print("[Step 5] Generating Visualization...")
    viz.plot_2025_predictions(grid_2025)

if __name__ == "__main__":
    main()
```
